```python
model_sum_columns = {
          
      "likesCount": 'sum',
      "retweetCount": 'sum',
      "sentimentScore" : 'sum',
      
      "tweetID" : 'count',
            
      "percentPrice_0" : 'mean',
      "percentPrice_1" : 'mean',
      "percentPrice_2" : 'mean',
      "percentPrice_3" : 'mean',
      "percentPrice_4" : 'mean'
}
```
