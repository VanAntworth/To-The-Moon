```python
model_sum_columns = {
          
      "likesCount": 'sum',
      "retweetCount": 'sum',
      "sentimentScore" : 'mean',
      
      "tweetID" : 'count',
            
      "percentPrice_0" : 'mean',
      "percentPrice_1" : 'mean',
      "percentPrice_2" : 'mean',
      "percentPrice_3" : 'mean',
      "percentPrice_4" : 'mean'
}
```

```python
model_average_columns = {
          
      "likesCount": 'mean',
      "retweetCount": 'mean',
      "sentimentScore" : 'mean',
      
      "tweetID" : 'count',
            
      "percentPrice_0" : 'mean',
      "percentPrice_1" : 'mean',
      "percentPrice_2" : 'mean',
      "percentPrice_3" : 'mean',
      "percentPrice_4" : 'mean'
}
```

```python
deltas_Twitter_df = df_deltas_restore[(df_deltas_restore["financeType"] == "twitter")]
deltas_Tesla_df = df_deltas_restore[(df_deltas_restore["financeType"] == "tesla")]
deltas_Doge_df = df_deltas_restore[(df_deltas_restore["financeType"] == "doge")]
```

```python
model_twitter_df_average = deltas_Twitter_df.groupby('startDate').agg(model_average_columns)
model_tesla_df_average = deltas_Tesla_df.groupby('startDate').agg(model_average_columns)
model_doge_df_average = deltas_Doge_df.groupby('startDate').agg(model_average_columns)
```
