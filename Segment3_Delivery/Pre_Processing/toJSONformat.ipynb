```python
#Fetch Tweets and Sentiment data - Join from Tweet and Sentiment tables
try:
    df_dbFinaceDeltaData = SqlConn.fetchFinanceDeltaPercents()
    print("Data transfer Done")
except BaseException as err:
    print(err)
```

```python
#Fetch Tweets and Sentiment data - Join from Tweet and Sentiment tables
try:
    df_dbFinaceTweetForecast = SqlConn.fetchFinanceTweetForecast()
    print("Data transfer Done")
except BaseException as err:
    print(err)
```

```python
df_dbFinaceDeltaData
```

```python
tweetsperDate = {}
financeTypes = ['tesla','doge','twitter']

for t in financeTypes:
    
    financeTypes_df = pd.DataFrame()
    tweetDates = []
    
    tweetsPerFinance = {}
    
    financeTypes_df = df_dbFinaceDeltaData[(df_dbFinaceDeltaData['financeType'] == t)]
    tweetDates = financeTypes_df['date'].unique().tolist()
    
    for date in tweetDates:
        
        tweetsperDate_df = pd.DataFrame()
        
        tweetsperDate_df = financeTypes_df[(financeTypes_df['date'] == date)]
        tweetIDs = tweetsperDate_df['tweetID'].tolist()
        
        tweetsPerFinance[date] = tweetIDs
        
    tweetsperDate[t] = tweetsPerFinance
```

```python
financeData7Days = {}

financeTypes = ['tesla','doge','twitter']

for t in financeTypes:
    
    financeTypes_df = pd.DataFrame()
    tweetIDs = []
    
    financePerTweet = {}
    
    financeTypes_df = df_dbFinaceTweetForecast[(df_dbFinaceTweetForecast['financeType'] == t)]
    tweetIDs = financeTypes_df['tweetID'].unique().tolist()
    
    for tID in tweetIDs:
        
        financePerTweet_df = pd.DataFrame()
        financeDates = []
        volumes = []
        adjustedClosedPrices = []
        datePositions = []
        financeInfo = {}

        
        financePerTweet_df = financeTypes_df[(financeTypes_df['tweetID'] == tID)].reset_index()
        
        tweetDate = financePerTweet_df['tweetDate'][0]
        financeDates = financePerTweet_df['financeDate'].tolist()
        volumes = financePerTweet_df['volume'].tolist()
        adjustedClosedPrices = financePerTweet_df['adjustedClose'].tolist()
        datePositions = financePerTweet_df['datePosition'].tolist()
        
        financeInfo = {'tweetDate':tweetDate ,'financeDates':financeDates, 'volumes':volumes, 'adjustedClosed':adjustedClosedPrices, 'datePositions': datePositions}
        
        financePerTweet[tID] = financeInfo
        
    financeData7Days[t] = financePerTweet
```
